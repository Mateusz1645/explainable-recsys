```python
# ── Load data and build features ──────────────────────────────────────
import sys
from pathlib import Path

PROJECT_ROOT = Path('/content/explainable-recsys')
sys.path.append(str(PROJECT_ROOT))

from src.eda.data_loader import load_data_tables
from src.eda.paths import get_file_map
from src.features.feature_engineering import build_all_features

file_map = get_file_map(PROJECT_ROOT)
data = load_data_tables(file_map)

user_feats, movie_feats, interact_feats = build_all_features(data)

print(user_feats.shape)
print(movie_feats.shape)
print(interact_feats.shape)
```

```python
# ── Inspect features ──────────────────────────────────────────────────
print('USER FEATURES:')
display(user_feats.head())

print('MOVIE FEATURES:')
display(movie_feats.head())

print('INTERACTION FEATURES:')
display(interact_feats.head())
```

```python
# ── Save to disk ──────────────────────────────────────────────────────
import os

os.makedirs('data/processed', exist_ok=True)

user_feats.to_parquet('data/processed/user_features.parquet', index=False)
movie_feats.to_parquet('data/processed/movie_features.parquet', index=False)
interact_feats.to_parquet('data/processed/interaction_features.parquet', index=False)

print('Saved to data/processed/')
```
